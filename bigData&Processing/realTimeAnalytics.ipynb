#### Start Spark Session

```python
# Initialize Spark Session
spark = SparkSession.builder.appName("RedditLiveAnalytics").getOrCreate()

# Only shows errors
spark.sparkContext.setLogLevel("ERROR")
```

#### Streaming Input Folder

```python
local_folder = "/home/hduser/Desktop/CA2/wheatherstreaming"
print("Streaming input folder created:", local_folder)
stop_api = False
```

#### Background Thread: Fetch API data continuously

```python
import requests, json, time, os
from datetime import datetime, timezone
import threading

api_key = "b49aa24061e313a458a86bd0d48a1512"
cities = ["London", "New York", "Tokyo", "Dublin"]

def fetch_stream():
    # Fetch continuously live API data and write JSON files
    global stop_api
    while not stop_api:
        for city in cities:
            url = f"http://api.openweathermap.org/data/2.5/weather?q={city}&appid={api_key}&units=metric"
            data = requests.get(url).json()
            
            timestamp = datetime.now(timezone.utc).strftime("%Y%m%d%H%M%S%f")
            filename = f"{local_folder}/weather_{city}_{timestamp}.json"
            
            with open(filename, "w") as f:
                json.dump(data, f)
                
            print("Saved:", filename)
            
        time.sleep(5) # fetch every 5 seconds    
                
# Start producer data in background
producer_thread = threading.Thread(target=fetch_stream)
producer_thread.daemon = True # allows notebook to stop it automatically
producer_thread.start()

print("API producer started...")
```

#### Define Schema and Read Live Stream

```python
from pyspark.sql.functions import col, current_timestamp, lit

schema = """
    coord STRUCT<lon: DOUBLE, lat: DOUBLE>,
    weather ARRAY<STRUCT<main: STRING, description: STRING>>,
    main STRUCT<temp: DOUBLE, humidity: DOUBLE>,
    wind STRUCT<speed: DOUBLE>,
    name STRING
"""


raw_data = spark.readStream.format("json").schema(schema).load("file://" + local_folder)

df = raw_data.select(
    col("name").alias("city"),
    
    # temperature
    col("main.temp").alias("temperature"),
    lit("%").alias("temperature_unit"),
    
    # humidity
    col("main.humidity").alias("humidity"),
    lit("%").alias("humidity_unit"),
    
    # wind
    col("wind.speed").alias("wind_speed"),
    lit("m/s").alias("wind_speed_unit"),
    
    col("weather")[0]["main"].alias("condition"),
    current_timestamp().alias("event_time")
)

df
```

#### Real-Time Analytics

```python
# Real-time alerts
low_temp_alerts = df.where("temperature < 10")
high_wind_alerts = df.where("wind_speed > 15")
```

```python
from pyspark.sql.functions import window, avg

# Real-time temperature averages (1 minute window)
temp_avg = df.groupby(
    window(col("event_time"), "1 minute"),
    col("city")
).agg(
    avg("temperature").alias("avg_temp")
)
```

#### Start the Streaming Queries

```python
# Temperature average
query_avg = temp_avg.writeStream.outputMode("update").format("memory").queryName("avg_temp_table").start()
```

```python
# Temperature alerts
query_temp_alert = low_temp_alerts.writeStream.outputMode("append").format("memory").queryName("low_temp_table").start()
```

```python
# Wind alerts
query_wind_alert = high_wind_alerts.writeStream.outputMode("append").format("memory").queryName("high_wind_table").start()
```

#### Stop the Streaming Queries and the Producer Thread

```python
for q in spark.streams.active:
    q.stop()
    
print("All Spark streaming queries stopped")

stop_api = True
try:
    producer_thread.join()
    print("API producer thread stopped")
except:
    print("Producer thread was already inactive")
```

#### Show Tables with the values collected

```python
# Average temperature
df_avg_temp = spark.sql("""
SELECT window.start AS window_start,
       window.end AS window_end,
       city,
       avg_temp
FROM avg_temp_table
ORDER BY window_start DESC, city
""")

df_avg_temp.show(df_avg_temp.count(), truncate=False)
```

```python
# Low temperature
df_low_temp = spark.sql("""
SELECT city, temperature, temperature_unit, condition, event_time 
FROM low_temp_table
ORDER BY event_time DESC
""")

df_low_temp.show(df_low_temp.count(), truncate=False)
```

```python
# High Winds
df_high_winds = spark.sql("""
SELECT city, wind_speed, wind_speed_unit, condition, event_time 
FROM high_wind_table
ORDER BY event_time DESC
""")

if df_high_winds.rdd.isEmpty():
    print("No high wind conditions detected at this time")
else:
    df_high_winds.show(df_high_winds.count(), truncate=False);

```
